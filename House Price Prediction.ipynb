### Apply feature selection to get top features having high dependency on target variable. Imported SelectKbest to  get k number independent variable with Mutual info regression. 
### Mutual info regression is measure of dependency of independent variable on contineous target variable.Higher its value, higher will be dependency of particular independent variable with target variable. 
     

```python
from sklearn.feature_selection import SelectKBest  
from sklearn.feature_selection import mutual_info_regression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(labels=['house price of unit area'],axis=1),df['house price of unit area'],test_size=0.3,random_state=4)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
top_five = SelectKBest(mutual_info_regression,k=5)
```

```python
top_five.fit(x_train,y_train)
```

```python
top_five.get_support()
```

```python
x_train.columns[top_five.get_support()]
```

```python
x_test=x_test[x_test.columns[top_five.get_support()]]
```

```python
x_train=x_train[x_train.columns[top_five.get_support()]]
```

```python
x_train
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
y_pred=LR.predict(x_test)
```

```python
y_test.tail()
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
```

```python
print('r2 score of the model is',score)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
MAE = mean_absolute_error(y_test,y_pred)
```

```python
MAE
```

```python
from sklearn.metrics import mean_absolute_percentage_error 
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```
